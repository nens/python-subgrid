```python
%matplotlib inline
import pandas
import matplotlib.pyplot as plt
import matplotlib.style
matplotlib.style.use('ggplot')
```

```python
df = pandas.read_csv('/Users/baart_f/Documents/3di.csv', sep=';').icol([1,2,3])
df.columns = ['type', 'count', 'file']
```

```python
import matplotlib.cm
import numpy as np
numbers = np.linspace(0,1, num=len(df.index))
np.random.shuffle(numbers)
colors = matplotlib.cm.Accent(numbers, alpha=1)
```

```python
summed = df.groupby('type').sum()
plt.subplots(figsize=(10,10))
_ = plt.pie(summed, labels=summed.index, colors=colors)
```
